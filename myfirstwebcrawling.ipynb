```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# requests 라이브러리를 사용하여 웹 페이지에서 HTML 데이터를 가지고 온다
req = requests.get('https://finance.naver.com/')
html = req.text
```

```python
print(html)
```

```python
# BeautifulSoup 객체 생성 후 HTML 데이터 파싱한다(HTML 코드 형태로 구분!!)
soup = BeautifulSoup(html, 'html.parser')
```

```python
# CSS 선택자를 활용하여 주요뉴스 추출
news = soup.select('.news_area a')
```

```python
title = []
url = []

for n in news:
    title.append(n.text)
    url.append(n['href'])

df = pd.DataFrame()
df['제목'] = title
df['URL'] = url

df
```
